## Accessing Data
Pandas allows us to access information using both slicing AND words

```python
pt_data.iloc[35] #slicing panda dataframe using column and row numbers
#dataframe_name.iloc[row_slice,col_slice]
```

```python
pt_data.iloc[35,:3]
```

```python
#quiz: use iloc function to select row 50 to the end (every other row) and cols 2-4
```

```python
pt_data.iloc[49::2,1:4]
```

```python
#to get data using words:
#dataframe_name["col_name"]
pt_data.columns #lists column/header names from file
```

```python
pt_data["AtomicMass"]
```

```python
pt_data[["AtomicMass","Symbol"]]
```

```python
pt_data[["Symbol","AtomicMass"]] #grabs data in order requested
```

```python
#third way to get information using panda
#loc function lets us use both column and row names
pt_data.loc[35,"Symbol"] #note that our rows are names with numbers
```

```python
#can do calcs on entire columns at once
pt_data["BoilingPoint"]
```

```python
#convert Kelvin values to celcius
pt_data["BoilingPoint"] - 273.15 #pandas applies mathematical operations to all elements of a columns (just like numpy arrays)
```

```python
pt_data["BoilingPointC"] = pt_data["BoilingPoint"] - 273.15 
#creates new col of data, appends to end of current table
```
